```python_programming_example
import re

# Example string
input_string = "We are learning regex with geeksforgeeks. Regex is very useful."

# Pattern to search for
pattern = "regex"

# Using re.search() to find the first occurrence of the pattern
match_obj = re.search(pattern, input_string, re.IGNORECASE)

if match_obj:
    # Get the entire matched substring
    print(f"Match found: {match_obj.group()}")

    # If you have capturing groups, you can access them like this:
    # print(f"Capturing group 1: {match_obj.group(1)}")
    # print(f"Capturing group 2: {match_obj.group(2)}")
else:
    print("Pattern not found in the string.")


Q.5. In the regex which created from the r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group zero cover?
Group 2? Group 1?

A.5. The pattern shown defines two capturing groups. Group 0 covers the entire match, and groups 1 and 2 cover the first and second parentheses, respectively.

Group 0 (Entire Match):
- Group 0 represents the entire matched substring.
- In this case, it covers the entire pattern: (\d\d\d)-(\d\d\d-\d\d\d\d).
- When you use .group(0) or simply .group(), it returns the entire matched string.

Group 1:
- Group 1 corresponds to the first capturing group, which is (\d\d\d).
- It captures three consecutive digits (a number).
- When you use .group(1), it returns the value of the first capturing group.

Group 2:
- Group 2 corresponds to the second capturing group, which is (\d\d\d-\d\d\d\d).
- It captures a sequence of three digits, a hyphen, and four more digits (e.g., “123-4567”).
- When you use .group(2), it returns the value of the second capturing group.

Q.6. In standard expression syntax, parentheses and intervals have distinct meanings. How can you tell
a regex that you want it to fit real parentheses and periods in python?

A.6. In Python regular expressions, parentheses and periods have special meanings. To treat them as literal characters (rather than special regex syntax), you can escape them using a backslash (\). Here’s how:

- Escaping Parentheses:
To match actual parentheses characters, use \\( to represent an opening parenthesis and \\) for a closing parenthesis.
For example, if you want to match the string “(hello)”, your regex pattern would be \\(hello\\).

- Escaping Periods (Dot):
The period (.) in regex matches any character except a newline.
To match a literal period, use \\. in your pattern.
For instance, to find the string “www.example.com”, your regex would be www\\.example\\.com.

Q.7. The findall() method returns a string list or a list of string tuples. What causes it to return one of
the two options?

A.7. The findall() method in pyhton depends on the pattern being used. Lets see different patterns.

1. No Capturing Groups:
    
    - If the pattern has no capturing groups, the findall() function returns a list of strings that match the whole pattern.
     - In this case, each item in the list represents a complete match.

     import re

     s = "black, blue and brown"

     pattern = r'bl\w+'

     matches = re.findall(pattern, s)

     print(matches)  # Output: ['black', 'blue']
     

2. Single Capturing Group:
- If the pattern has one capturing group, the findall() function returns a list of strings that match the group.
- The capturing group is specified using parentheses.   

import re

s = "black, blue and brown"

pattern = r'bl(\w+)'

matches = re.findall(pattern, s)

print(matches)  # Output: ['ack', 'ue']

3. Multiple Capturing Groups:
- If the pattern has multiple capturing groups, the findall() function returns a list of tuples.
- Each tuple represents a complete match.
- Each tuple contains the strings that match the individual groups.

import re

s = "black, blue and brown"

pattern = r'(bl(\w+))'

matches = re.findall(pattern, s)

print(matches)  # Output: [('black', 'ack'), ('blue', 'ue')]

4. Non-Capturing Groups:
- Non-capturing groups (specified using (?:...)) do not affect the form of the return result.
- They are not included in the output of findall().

In summary:
- No capturing groups: List of whole pattern matches.
- Single capturing group: List of group matches.
- Multiple capturing groups: List of tuples with group matches.







Q.8. In standard expressions, what does the | character mean?

A.8. The | character means "or" in the context of a boolean expression. It serves as an alternation operator.

    1. Alternation:
- The | character allows you to specify multiple alternatives within a pattern.
- It acts like an OR operator, allowing the regex engine to match any of the alternatives.
- For example, the pattern cat|dog will match either “cat” or “dog” in the input string.

import re

s = "I have a cat and a dog."

pattern = r"cat|dog"

matches = re.findall(pattern, s)

print(matches)  # Output: ['cat', 'dog']

2.Usage:
- You can use parentheses to group alternatives together.
- For instance, (cat|dog) will match either “cat” or “dog” as a whole.


Q.10. In regular expressions, what is the difference between the + and * characters?

A.10. The + and * characters serve as quantifiers to control the repetition of the preceding expression. The + character means that the preceding expression must occur one or more times. The * character means that the preceding expression must occur zero or more times.
+ requires at least one occurrence.
* allows zero or more occurrences.

Examples

- Suppose we have the input string “abcaaaab”.
- The pattern a+b will match “aaa” (one or more “a” characters).
- The pattern a*b will match “a” (zero or more “a” characters).

Q.11. What is the difference between {4} and {4,5} in regular expression?

A.11. {4}:
- The expression {4} specifies that the preceding element must occur exactly 4 times.
- It is a fixed quantifier that matches exactly the specified number of occurrences.

For example:
- The pattern a{4} will match “aaaa” but not “aaa” or “aaaaa”.

{4,5}:
- The expression {4,5} specifies that the preceding element must occur between 4 and 5 times, inclusive.
- It is a range quantifier that allows flexibility in the number of occurrences.

For example:
- The pattern a{4,5} will match “aaaa” or “aaaaa” but not “aaa” or “aaaaaa”.

Examples:

- Suppose we have the input string “abcaaaab”.
- The pattern a{4} will match “aaaa”.
- The pattern a{4,5} will match both “aaaa” and “aaaaa”.

Q.12. What do you mean by the \d, \w, and \s shorthand character classes signify in regular
expressions?

A.12. \d (Digit Character Class):

- The \d represents the digit character class.
- It matches any single digit (0-9).

For example:
- The pattern \d will match any digit in the input string.

 \w (Word Character Class):
- The \w represents the word character class.
- It matches any alphanumeric character (letters, digits, or underscores).
- Specifically, it includes characters from the ranges A-Z, a-z, 0-9, and the underscore (_).
For example:
- The pattern \w will match any word character in the input string.

\s (Whitespace Character Class):
- The \s represents the whitespace character class.
- It matches any whitespace character, including spaces, tabs, newlines, carriage returns, form feeds, and vertical tabs.

For example:
- The pattern \s will match any whitespace character in the input string.



Q.13. What do means by \D, \W, and \S shorthand character classes signify in regular expressions?

A.13. \D (Non-Digit Character Class):
- The \D represents the non-digit character class.
- It matches any single character that is not a digit (0-9).

For example:
- The pattern \D will match any non-digit character in the input string.

\W (Non-Word Character Class):
- The \W represents the non-word character class.
- It matches any character that is not an alphanumeric character (letters, digits, or underscores).
- Specifically, it includes characters other than A-Z, a-z, 0-9, and the underscore (_).

For example:
- The pattern \W will match any non-word character in the input string.

\S (Non-Whitespace Character Class):

- The \S represents the non-whitespace character class.
- It matches any character that is not a whitespace character (space, tab, newline, etc.).

For example:
- The pattern \S will match any non-whitespace character in the input string.

Q.14. What is the difference between .*? and .*?

A.14. *? (Lazy or Non-Greedy Match):
- The .*? is a non-greedy or lazy match.
- It matches as few characters as possible while still allowing the rest of the pattern to match.

For example:
- If you have the input string “abc123def456”, the pattern a.*?d will match “abc1d” (matching the shortest substring between “a” and “d”).

.* (Greedy Match):
- The .* is a greedy match.
- It matches as many characters as possible while still allowing the rest of the pattern to match.

For example:
- If you have the input string “abc123def456”, the pattern a.*d will match “abc123def” (matching the longest substring between “a” and “d”).

Usage:
- Use .*? when you want to match the shortest possible substring.
- Use .* when you want to match the longest possible substring.

Q.15.What is the syntax for matching both numbers and lowercase letters with a character class?

A.15.We  can use classes to match both numbers and lowercase letters. Examples:

1. To match any alphanumeric character (letters or digits), you can use the \w shorthand character class. It includes lowercase letters, uppercase letters, digits (0-9), and the underscore (_). Equivalent to [a-zA-Z0-9_].
2. If you specifically want to match only lowercase letters and digits, you can use the character class [a-z0-9].

E.g.

import re

s = "abc123def456"

pattern = r"\w{8}"  # Matches any 8 alphanumeric characters

matches = re.findall(pattern, s)

print(matches)  # Output: ['abc123de', 'f456']

E.g. 

import re

s = "abc123def456"

pattern = r"[a-z0-9]{8}"  # Matches 8 lowercase letters or digits

matches = re.findall(pattern, s)

print(matches)  # Output: ['abc123de', 'f456']



Q.16. What is the procedure for making a normal expression in regex case insensitive?

A.16. 
 1. Using re.compile() with the re.IGNORECASE Flag:
You can compile a regular expression pattern using re.compile() and pass the re.IGNORECASE flag as the second argument.
This flag ensures that the pattern matches regardless of letter case.

2. Directly Using re.search(), re.match(), or re.sub() with the re.IGNORECASE Flag:

You can directly use re.search(), re.match(), or re.sub() with the re.IGNORECASE flag.

3. Using Inline Flags with (?i):
You can perform case-insensitive searches using search() or match() without explicitly using the re.IGNORECASE flag.

Q.17. What does the . character normally match? What does it match if re.DOTALL is passed as 2nd
argument in re.compile()?

A.17. The . character in Python regular expressions (regex) has different behavior depending on the context:

Normal Behavior (Without re.DOTALL):

- By default, the . matches any character except a newline.
- For example, if you use the pattern "he..o", it will match strings like “hello”, “heXo”, and “he!o”, but not “he\no” (where \n represents a newline).
- Essentially, the dot matches any character except newline characters.

With re.DOTALL (re.S) Flag:
- When you pass the re.DOTALL (or re.S) flag as the second argument to re.compile(), the behavior of the dot changes.
- With this flag, the dot matches any character, including newline characters.
- For example, if you use the pattern "he..o" with the re.DOTALL flag, it will match strings like “he\no” as well.

Additionally:

- Without the flag, the dot matches any character except newline.
- With the flag, the dot matches any character, including newline 1234.

Q.18.If numReg = re.compile(r'\d+'), what will numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4
hen') return in python?

A.18. Replace all digits with ‘X’ in the given string using regex in Python.

- If we have a compiled regex pattern numReg = re.compile(r'\d+'), and we use the sub() method with the replacement string 'X' on the input string '11 drummers, 10 pipers, five rings, 4 hens', the result will be:

- 'X drummers, X pipers, five rings, X hens'

- In this output, all occurrences of digits (like 11, 10, and 4) have been replaced with 'X'.

Q.19.What does passing re.VERBOSE as the 2nd argument to re.compile() allow to do?

A.19. The re.VERBOSE flag, also known as re.X, is a powerful feature in Python’s regular expressions (regex). When you pass this flag as the second argument to re.compile(), it allows you to write more readable and well-organized regex patterns by providing the following benefits:

Improved Readability:
- The re.VERBOSE flag allows you to write complex regex patterns in a more visually separated and organized manner.
- You can add comments within the pattern to explain different sections or logical components.
- Whitespace within the pattern is ignored (except in specific cases), making it easier to format and structure your regex.

Whitespace Ignoring:
- Within the pattern, whitespace (spaces, tabs, and newlines) is ignored unless it is:
- Inside a character class (e.g., [a-z]).
- Preceded by an unescaped backslash (e.g., \\s for matching whitespace).
- Within tokens like *?, (?:), or (?P.
- This allows you to add indentation, line breaks, and comments without affecting the actual matching behavior.

Comments and Clarity:
- we can use inline comments after the # character to break down complex regex into multiple lines.
- Comments help explain the purpose of each part of the pattern, making it easier for others (and your future self) to understand.

E.g. 

#### Without VERBOSE
regex_email = re.compile(r'^([a-z0-9_\\.-]+)@([0-9a-z\\.-]+)\\.([a-z\\.]{2,6})$', re.IGNORECASE)

#### Using VERBOSE
regex_email = re.compile(r"""

    ^([a-z0-9_\\.-]+)  # local Part

    @                 # single @ sign

    ([0-9a-z\\.-]+)   # Domain name

    \\.\              # single Dot .

    ([a-z]{2,6})$     # Top level Domain
    
""", re.VERBOSE | re.IGNORECASE)


Q.20. How would you write a regex that match a number with comma for every three digits? It must
match the given following:
'42'
'1,234'
'6,369,745'

but not the following:
'12,34,567' (which has only two digits between the commas)
'1234' (which lacks commas).

A.20. import re

pattern = r'^\d{1,3}(,\d{3})*$'

#### Test cases
test_strings = [

    '42',

    '1,234',

    '6,369,745',

    '12,34,567',  # Should not match

    '1234'        # Should not match
]

for test_string in test_strings:

    if re.match(pattern, test_string):

        print(f"'{test_string}' matches the pattern.")

    else:
    
        print(f"'{test_string}' does not match the pattern.")


- ^: Matches the start of the string.

- \d{1,3}: Matches 1 to 3 digits (the first set of digits).

- (,\d{3})*: Matches zero or more occurrences of a comma followed by exactly 3 digits (subsequent sets of digits).

- $: Matches the end of the string.

Q.21.How would you write a regex that matches the full name of someone whose last name is
Watanabe? You can assume that the first name that comes before it will always be one word that
begins with a capital letter. The regex must match the following:

'Haruto Watanabe'

'Alice Watanabe'

'RoboCop Watanabe'

but not the following:

'haruto Watanabe' (where the first name is not capitalized)

'Mr. Watanabe' (where the preceding word has a nonletter character)

Watanabe' (which has no first name)

'Haruto watanabe' (where Watanabe is not capitalized)

A.21.
  import re

pattern = r'^[A-Z][a-z]* Watanabe$'

# Test cases
test_names = [

    'Haruto Watanabe',

    'Alice Watanabe',
    'RoboCop Watanabe',
    'haruto Watanabe',   # Should not match
    'Mr. Watanabe',      # Should not match
    'Watanabe',          # Should not match
    'Haruto watanabe'    # Should not match
]

for name in test_names:

    if re.match(pattern, name):
        print(f"'{name}' matches the pattern.")
    else:
        print(f"'{name}' does not match the pattern.")

Explanation of the pattern:

^: Matches the start of the string.

[A-Z]: Matches an uppercase letter (the first letter of the first name).

[a-z]*: Matches zero or more lowercase letters (the remaining letters of the first name).

Watanabe: Matches the last name “Watanabe.”

$: Matches the end of the string.

The pattern ensures that the first name starts with an uppercase letter and is followed by lowercase letters, and the last name is specifically “Watanabe.”        


Q.22. How would you write a regex that matches a sentence where the first word is either Alice, Bob,
or Carol; the second word is either eats, pets, or throws; the third word is apples, cats, or baseballs;
and the sentence ends with a period? This regex should be case-insensitive. It must match the
following:

'Alice eats apples.'

'Bob pets cats.'

'Carol throws baseballs.'

'Alice throws Apples.'

'BOB EATS CATS.'


but not the following:

'RoboCop eats apples.'

'ALICE THROWS FOOTBALLS.'

'Carol eats 7 cats.'

A.22. 

import re

pattern = r'^(Alice|Bob|Carol) (eats|pets|throws) (apples|cats|baseballs)\.$'

# Test cases

test_sentences = [

    'Alice eats apples.',
    'Bob pets cats.',
    'Carol throws baseballs.',
    'Alice throws Apples.',
    'BOB EATS CATS.',       # Should match (case-insensitive)
    'RoboCop eats apples.', # Should not match
    'ALICE THROWS FOOTBALLS.', # Should not match
    'Carol eats 7 cats.'    # Should not match
]

for sentence in test_sentences:

    if re.match(pattern, sentence, re.IGNORECASE):
        print(f"'{sentence}' matches the pattern.")
    else:
        print(f"'{sentence}' does not match the pattern.")


^: Matches the start of the string.

(Alice|Bob|Carol): Matches any of the specified first names.

(eats|pets|throws): Matches any of the specified verbs.

(apples|cats|baseballs): Matches any of the specified objects.

\.: Matches a period (escaped because it has a special meaning in regex).

$: Matches the end of the string.
